```python
df.rename(columns={'Pago':'Pago_M0', 'Fecha_pago': 'Fecha_pago_M0', 'Fecha_limite_pago':'Fecha_limite_pago_M0'}, inplace=True)
```

```python
for k in range(6, -1, -1):
    col = 'dias_deudados_m' + str(k)
    fecha_limite_col = 'Fecha_limite_pago_M' + str(k)
    
    # Initialize column with None
    df[col] = None
    
    # For each possible payment month that could cover month k
    for i in range(0, k+1):
        pago_col = 'Pago_M' + str(k - i)
        fecha_pago_col = 'Fecha_pago_M' + str(k - i)
        
        # Find rows where:
        # 1. This payment is > 0
        # 2. The dias_deudados column is still None (meaning no earlier payment found)
        # 3. Both date columns contain valid dates
        mask = (
            (df[pago_col] > 0) & 
            df[col].isna() & 
            pd.notna(df[fecha_limite_col]) & 
            pd.notna(df[fecha_pago_col])
        )
        
        # Calculate days difference only for the matching rows
        if mask.any():
            df.loc[mask, col] = (
                df.loc[mask, fecha_limite_col] - 
                df.loc[mask, fecha_pago_col]
            ).dt.days
```

```python
df['dias_deudados_m6'].fillna(0, inplace=True)
df['dias_deudados_m5'].fillna(df['dias_deudados_m6'] - 30, inplace=True)
df['dias_deudados_m4'].fillna(df['dias_deudados_m5'] - 30, inplace=True)
df['dias_deudados_m3'].fillna(df['dias_deudados_m4'] - 30, inplace=True)
df['dias_deudados_m2'].fillna(df['dias_deudados_m3'] - 30, inplace=True)
df['dias_deudados_m1'].fillna(df['dias_deudados_m2'] - 30, inplace=True)
df['dias_deudados_m0'].fillna(df['dias_deudados_m1'] - 30, inplace=True)
```

# Variable objetivo Mes i


```python
df.rename(columns={'Pago_minimo':'Pago_minimo_M0'}, inplace=True)
```

```python
for k in range(6, 0, -1):
    col = 'Variable_Objetivo_m' + str(k)    
    # Initialize column with None
    df[col] = None
    
    # For each possible payment month that could cover month k
    for i in range(0, k+1):
        pago_col = 'Pago_M' + str(k - i)
        pago_min_col = 'Pago_minimo_M' + str(k - i)

        
        # Find rows where:
        # 1. This payment is > pago_min_col
        mask_true = (
            (df[pago_col] > df[pago_min_col]) & 
            df[col].isna()
        )
        
        mask_false = (
            (df[pago_col] <= df[pago_min_col]) |
            df[pago_min_col].isna() &
            df[col].isna()
        )
        
        # Calculate Variable objetivo
        if mask_true.any():
            df.loc[mask_true, col] = 1
            
        df.loc[mask_false, col] = 0
```
